```python
import glob, os
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = []
data_2 = []
for file in glob.glob("/home/nibraas_a_khan/Working-Memory-Temporal-Difference/combined_model/all_stats/static_t_non_obs_stats/newStats/static/*"):
    f = open(file)
    data += [float(f.read().splitlines()[0])]
for file in glob.glob("/home/nibraas_a_khan/Working-Memory-Temporal-Difference/combined_model/all_stats/dynamic_t_non_obs_stats/newStats/reset/*"):
    f = open(file)
    data_2 += [float(f.read().splitlines()[0])]
sns.kdeplot(data, label="Static", shade=True)
sns.kdeplot(data_2, label="Dynamic", shade=True)
plt.legend()
```
